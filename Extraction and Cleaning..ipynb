```python
# importing libraries and Setting Option

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from bs4 import BeautifulSoup
import requests
import re
```

```python
data19 = pd.read_csv('players_2019.csv')
data20 = pd.read_csv('players_2020.csv')
data21 = pd.read_csv('players_2021.csv')
data22 = pd.read_csv('players_2022.csv')
```

```python
data19.head()
```

```python
data19['Team'] = data19['Team & Contract'].apply(lambda val: val.split('\n')[0])
data19['Contract'] = data19['Team & Contract'].apply(lambda val: val.split('\n')[1])
data19 = data19.drop(['Team & Contract'], axis=1)

data20['Team'] = data20['Team & Contract'].apply(lambda val: val.split('\n')[0])
data20['Contract'] = data20['Team & Contract'].apply(lambda val: val.split('\n')[1])
data20 = data20.drop(['Team & Contract'], axis=1)
                                             
data21['Team'] = data21['Team & Contract'].apply(lambda val: val.split('\n')[0])
data21['Contract'] = data21['Team & Contract'].apply(lambda val: val.split('\n')[1])
data21 = data21.drop(['Team & Contract'], axis=1)
    
data22['Team'] = data22['Team & Contract'].apply(lambda val: val.split('\n')[0])
data22['Contract'] = data22['Team & Contract'].apply(lambda val: val.split('\n')[1])
data22 = data22.drop(['Team & Contract'], axis=1)                                             
```

```python
data19['Height'] = data19['Height'].apply(lambda val: int(val.strip('cm')))
data20['Height'] = data20['Height'].apply(lambda val: int(val.strip('cm')))
data21['Height'] = data21['Height'].apply(lambda val: int(val.strip('cm')))
data22['Height'] = data22['Height'].apply(lambda val: int(val.strip('cm')))
```

```python
data19['Weight'] = data19['Weight'].apply(lambda val: int(val.strip('kg')))
data20['Weight'] = data20['Weight'].apply(lambda val: int(val.strip('kg')))
data21['Weight'] = data21['Weight'].apply(lambda val: int(val.strip('kg')))
data22['Weight'] = data22['Weight'].apply(lambda val: int(val.strip('kg')))
```

```python
def convert(val):
    val = val.strip().strip('€')
    if 'M' in val:
        return float(val.strip('M')) * 1000000
    elif 'K' in val:
        return float(val.strip('K')) * 1000
    
data19['Value'] = data19['Value'].apply(convert)
data20['Value'] = data20['Value'].apply(convert)
data21['Value'] = data21['Value'].apply(convert)
data22['Value'] = data22['Value'].apply(convert)
```
